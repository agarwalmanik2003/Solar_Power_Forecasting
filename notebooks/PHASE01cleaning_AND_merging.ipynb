```python
import pandas as pd

def process_plant(gen_file, weather_file):
    gen = pd.read_csv(gen_file)
    weather = pd.read_csv(weather_file)

    gen["DATE_TIME"] = pd.to_datetime(gen["DATE_TIME"])
    weather["DATE_TIME"] = pd.to_datetime(weather["DATE_TIME"])

    gen = gen.drop_duplicates()
    weather = weather.drop_duplicates()

    gen = gen[gen["AC_POWER"] > 0]

    df = pd.merge(gen, weather, on="DATE_TIME")

    df["hour"] = df["DATE_TIME"].dt.hour
    df["day"] = df["DATE_TIME"].dt.day
    df["month"] = df["DATE_TIME"].dt.month

    return df

# -------- PROCESS BOTH PLANTS ----------

p1 = process_plant("Plant_1_Generation_Data.csv",
                   "Plant_1_Weather_Sensor_Data.csv")

p2 = process_plant("Plant_2_Generation_Data.csv",
                   "Plant_2_Weather_Sensor_Data.csv")

# -------- MERGE BOTH INTO ONE ----------

final_df = pd.concat([p1, p2])

# SAVE

final_df.to_csv("final_cleaned_dataset.csv", index=False)

print("Phase-1 done. Final dataset saved.")
print(final_df.shape)
```

SUMMARY OF PHASE01

In Phase-1, we loaded all four plant power and weather CSV files, converted the DATE_TIME column to proper datetime format, removed duplicates and zero-power (night) rows, and merged generation data with corresponding weather readings. We then engineered time-based features (hour, day, month), repeated the same cleaning for both plants, combined them into one dataset, and saved the final ML-ready file. The result was a single clean dataset — final_cleaned_dataset.csv — containing merged power + weather data with time features for accurate model training.
